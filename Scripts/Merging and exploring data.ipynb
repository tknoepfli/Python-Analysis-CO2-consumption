# 3. Wrangling data

```python
df2=pd.wide_to_long(df_hdi, stubnames=['HumanDevelopmentIndex', 'GrossNationalIncomePerCapita'], i=['Country'], j='Year', suffix='\d+', sep='_')
```

```python
df2.head(25)
```

```python
df2=df2.sort_values(['Country', 'Year'], ascending=[True,True])
```

```python
df2.head(25)
```

```python
df2.shape
```

# 4. Merging data

```python
df_left=pd.merge(df2, df_co2, how='left', on=['Country', 'Year'])
```

```python
df_left.shape
```

```python
df_hdi_co2_happy = pd.merge(df_left, df_happy, how='left', on=['Country', 'Year'])
```

```python
df_hdi_co2_happy.shape
```

```python
df_all = pd.merge(df_hdi_co2_happy, df_categorical, how='left', on=['Country'])
```
